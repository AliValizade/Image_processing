```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('input/home.png', 0)
plt.imshow(img, cmap='gray')
```

```python
_, thresh_img = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
thresh_img = 255 - thresh_img
plt.imshow(thresh_img, cmap='gray')
```

```python
# contours, _ = cv.findContours(thresh_img, cv.RETR_LIST, cv.CHAIN_APPROX_NONE)
rows, cols = img.shape
contours = []
for i in range(rows):
    for j in range(cols):
        if thresh_img[i, j] == 255:
            if thresh_img[i-1, j] == 0 or thresh_img[i+1, j] == 0 or thresh_img[i, j-1] == 0 or thresh_img[i, j+1] == 0:
                contours.append([(j, i)])
contours = np.array(contours)
cv.drawContours(img, contours, -1, (255, 255, 255), 4)
    
plt.imshow(img, cmap='gray')
```
