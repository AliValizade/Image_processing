```python
import cv2
import matplotlib.pyplot as plt
```

# Object detection (PUMA's LOGO)

```python
puma = cv2.imread('input/PUMA.jpg', 0)
_, thresh_puma = cv2.threshold(puma, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh_puma, cmap='gray')
```

```python
puma = 255 - thresh_puma
plt.imshow(puma, cmap='gray')
```

```python
puma_contours, _ = cv2.findContours(puma, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
for c in puma_contours:
    if len(c) > 100:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(puma, (x, y), (x+w, y+h), (255, 255, 255), 2)
plt.imshow(puma, cmap='gray')
```

# Object detection (COINS)

```python
coins = cv2.imread('input/coins.jpg', 0)
plt.imshow(coins, cmap='gray')
```

# Object detection (Wolf)

```python
wolf = cv2.imread('input/wolf.jpg', 0)
# plt.imshow(wolf, cmap='gray')
```

```python
_, thresh_wolf = cv2.threshold(wolf, 130, 255, cv2.THRESH_BINARY)
# plt.imshow(thresh_wolf, cmap='gray')
```

```python
wolf_contours, _ = cv2.findContours(thresh_wolf, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
for c in wolf_contours:
    if len(c) > 200:
        x, y, w, h = cv2.boundingRect(c)                        
        cv2.rectangle(wolf, (x, y), (x+w, y+h), (255, 0, 0), 3)

plt.imshow(wolf, cmap='gray')
```
