```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('input/young-couple.jpg')
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(imgRGB)
```

```python
imgHSV = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```

```python
h, s ,v = cv.split(imgHSV)
rows, cols = h.shape
```

```python
plt.imshow(v, cmap='gray')
```

```python
for i in range(rows):
    for j in range(cols):
        if 40 < h[i, j] < 80 and s[i, j] > 100:
            h[i, j] += 70
```

```python
imgResult = cv.merge((h, s, v))
imgResult = cv.cvtColor(imgResult, cv.COLOR_HSV2RGB)
plt.imshow(imgResult)
```
