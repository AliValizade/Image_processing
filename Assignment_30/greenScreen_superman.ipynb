```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('input/superman.jpg')
imgRGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(imgRGB)
```

```python
bg = cv.imread('input/sky1.jpg')
```

```python
imgHSV = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```

```python
h, s, v = cv.split(imgHSV)
row, col = h.shape
```

```python
bg = cv.resize(bg, (col, row))
```

```python
for i in range(row):
    for j in range(col):
        if 45 < h[i, j] < 75:
            img[i, j] = bg[i, j]
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv.cvtColor(img, cv.COLOR_RGB2BGR)
cv.imwrite('output/fly-superman.jpg', img)
```
