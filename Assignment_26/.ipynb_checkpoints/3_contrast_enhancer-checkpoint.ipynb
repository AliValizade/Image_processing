```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input/original-statue.jpg')
plt.imshow(img, cmap='gray')
```

```python
# convert color image from BGR to LAB (L: lightness(0:100)  A: Green-Red(-128:127)  B: Blue-Yellow(-128:127) )
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
# Split the channels
l, a, b = cv2.split(lab)
```

```python
# Apply CLAHE to the L-channel
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl = clahe.apply(l)
```

```python
# Merge the CLAHE enhanced L-channel with the a and b channel
limg = cv2.merge((cl,a,b))
```

```python
# Convert back to BGR color space
new_img = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)
plt.imshow(new_img, cmap='gray')
```

```python
# Show the result
cv2.imshow('Original', img)
cv2.imshow('Enhanced Contrast', new_img)
cv2.imwrite('output/enhanced_contrast_statue.jpg', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
