```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
logo = np.ones((650, 1320, 3), dtype=np.uint8) * 85
plt.imshow(logo)
```

```python
# red
logo[210:310, 210:310] = (246, 83, 20)
# green
logo[210:310, 320:420] = (124, 187, 0)
# blue
logo[320:420, 210:310] = (0, 161, 241)
# yellow
logo[320:420, 320:420] = (255, 187, 0)
```

```python
logo = cv.putText(logo, 'Microsoft', (470, 350), cv.FONT_HERSHEY_DUPLEX, 4, (255, 255, 255), 10)
```

```python
plt.imshow(logo)
```

```python
logo = cv.cvtColor(logo, cv.COLOR_RGB2BGR)
cv.imwrite('output/m-logo.jpg', logo)
```
