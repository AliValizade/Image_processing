```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
carrot = cv.imread('input/carrot.jpg')
carrot = cv.cvtColor(carrot, cv.COLOR_BGR2RGB)
plt.imshow(carrot)
```

```python
B,G,R=cv.split(carrot)
```

```python
R_new=G
G_new=G
B_new=R

carrot_green=cv.merge((B_new,G_new,R_new))
carrot_green=cv.cvtColor(carrot_green,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(carrot_green)
```
